TASK 2 : #1} Count the number of trips for each business in each month. Draw the histogram with 'business-month' on the x-axis and trip counts on the y-axis. For example, assume if the number of trips for Uber in January 

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings

# Reading csv file
df1 = pd.read_csv('task2_output/trips_per_business_month.csv/part-00000-c80b57de-4adc-48ee-84d6-0c7d1158509a-c000.csv')
df1 = df1.sort_values(by=['business','month'])

# Filtering uber and lyft business in separate data frames
df1_uber = df1[df1['business'] == 'Uber'].copy()
df1_lyft = df1[df1['business'] == 'Lyft'].copy()

# Creating the 'business-month' string for the x-axis for uber and lyft 

df1_uber['business_month'] = df1_uber['business'] + '-' + df1_uber['month'].astype(str)
df1_lyft['business_month'] = df1_lyft['business'] + '-' + df1_lyft['month'].astype(str)

# Plot the histogram for uber
plt.figure(figsize=(6, 4))
plt.bar(df1_uber['business_month'], df1_uber['trip_count'], color='orange')

plt.xlabel('Business-Month')
plt.ylabel('Trip Count')
plt.title('Number of Trips for Uber Each Month')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust the padding between and around subplots
plt.show()

# Plot the histogram for lyft
plt.figure(figsize=(6, 4))
plt.bar(df1_lyft['business_month'], df1_lyft['trip_count'], color='purple')

plt.xlabel('Business-Month')
plt.ylabel('Trip Count')
plt.title('Number of Trips for Lyft Each Month')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust the padding between and around subplots
plt.show()
```
